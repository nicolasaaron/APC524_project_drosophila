```python
#%matplotlib inline

import skimage
import skimage.io
import skimage.exposure
import skimage.morphology
import skimage.segmentation
import skimage.viewer
import skimage.measure

#better to write like
from skimage import io, exposure, morphology, segmentation, viewer, measure, color


import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage

filename = "C1-WT-9.png"
imflat = skimage.io.imread(filename,plugin='matplotlib')


print('image shape:',np.shape(imflat))
print('image date type:',imflat.dtype)
print('image scale', imflat.max(), imflat.min())

fig, ax = plt.subplots(figsize=(8, 8))
ax.axis('off')
tmp = ax.imshow(imflat,"gray")
plt.show()

imflatgray = color.rgb2gray(imflat.copy())

fig, ax = plt.subplots(figsize=(8, 8))
ax.axis('off')
ax.imshow(imflatgray,"gray")
plt.show()
```

```python
# test from skimage canny

from skimage import feature

edges1 = feature.canny(imflatgray)
edges2 = feature.canny(imflatgray, sigma=0.01)

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(8,8),sharex = True, sharey = True)

ax1.imshow(edges1, cmap = plt.cm.gray)
ax1.axis('off')
ax1.set_title('canny filter, $\sigma=1$', fontsize=20)

ax2.imshow(edges2, cmap = plt.cm.gray)
ax2.axis('off')
ax2.set_title('canny filter, $\sigma=3$', fontsize=20)
```

```python
# finding countours using opencv

import cv2

im = cv2.imread(filename)
print('data type of im', im.dtype)
imgray = cv2.cvtColor(im.copy(), cv2.COLOR_BGR2GRAY)


# plot figures
fig,(ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16,16),sharex = True, sharey = True)
ax1.imshow(im,'gray')
ax2.imshow(imgray,'gray')
ax1.axis('off')
ax1.set_title('origin', fontsize=10)
ax2.axis('off')
ax2.set_title('grayscale', fontsize=10)

# compute masked image (>=threshold) and detect contours
ret,thresh = cv2.threshold(imgray,8,255,cv2.THRESH_BINARY)
_,contours, hierarchy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

for contour in contours:
    area = cv2.contourArea(contour)        
    if area > 1000:
        print('dim contour i with area >1000',np.shape(contour))
        print('area of contour i',area)
        contour_image= cv2.drawContours(im.copy(), [contour], -1, (255,255,255), 3)
        contour_external = contour;

#figures for the masked image and the external contour
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16,16),sharex = True, sharey = True)
ax1.imshow(thresh,'gray')
ax1.axis('off')
ax1.set_title('threshold >=8', fontsize=10)
ax2.imshow(contour_image,'gray')
ax2.axis('off')
ax2.set_title('contour external', fontsize=10)


# find convex hull and polyapprox curve for the external contour
hull = cv2.convexHull(contour_external)
epsilon = 0.005 * cv2.arcLength(contour_external, True)
polyapprox = cv2.approxPolyDP(contour_external, epsilon,True)
print('dim convex hull', np.shape(hull))
print('dim poly approx',np.shape(polyapprox))

# figures for convex hull and polyapprox curve
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16,16),sharex = True, sharey = True)
image1 = cv2.drawContours(im.copy(), [hull], -1, (255,255,255), 3)
ax1.imshow(image1,'gray')
ax1.axis('off')
ax1.set_title('convex hull', fontsize=10)
image2 = cv2.drawContours(im.copy(), [polyapprox],-1,(255,255,255),3)
ax2.imshow(image2,'gray')
ax2.axis('off')
ax2.set_title('contour poly approx', fontsize=10)
```

```python
# convex hull detection using skimage with otsu method and mannual threshold

binary_im= imflatgray.copy()
print('image gray scale', binary_im.max(), binary_im.min())

threshold = 0.03
binary_im[imflatgray <= threshold] = 0.0
binary_im[imflatgray > threshold]  = 1.0

threshold = skimage.filters.threshold_otsu(imflatgray)
mask = imflatgray >= threshold
print('otsu threshold:', threshold)

hull_1= skimage.morphology.convex_hull_image(binary_im)
print("dim hull from binary detection:", np.shape(hull))
hull_2= skimage.morphology.convex_hull_image(mask)


fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(16,16), sharex=True, sharey=True)
ax1.imshow(binary_im,'gray')
ax1.set_title('binary image, threshold = 0.03 ', fontsize = 10)
ax2.imshow(hull_1, 'gray')
ax2.set_title('convex hull from binary image, threshold =0.03', fontsize = 10)

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(16,16), sharex=True, sharey=True)
ax1.imshow(mask, 'gray')
ax1.set_title('mask, threshold={:f}'.format(threshold), fontsize=10)
ax2.imshow(hull_2, 'gray')
ax2.set_title('convex hull from mask, threshold={:f}'.format(threshold), fontsize=10)
```

```python
#contour detetection with skimage.measure
threshold = skimage.filters.threshold_otsu(imflatgray)
print("threshold otsu:", threshold)

# contour detection using  find_contours
contours = measure.find_contours(imflatgray, threshold)
print('number of contours:', len(contours))
print('external contour, number of data points:', np.shape(contours[0]))

plt.figure(figsize=(24,24))
fig = plt.imshow(imflatgray,'gray')
for contour in contours:
    plt.plot(contour[:,1], contour[:,0], linewidth=2)
#plt.gca().invert_yaxis()
plt.show()


plt.figure(figsize=(24,24))
contour_ext = contours[0]
plt.imshow(imflatgray,'gray')
plt.plot(contour_ext[:,1], contour_ext[:,0], linewidth=1)
plt.show()


# contour detection using segmentation

mask = imflatgray >= threshold
clean_border = segmentation.clear_border(mask).astype(np.int)
cell_edges = segmentation.mark_boundaries(imflatgray, clean_border)

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(14,14), sharex=True, sharey=True)
ax1.imshow(clean_border, 'gray')
ax1.set_title('clean border', fontsize=10)
ax2.imshow(cell_edges, 'gray')
ax2.set_title('cell edges', fontsize=10)
plt.tight_layout()
plt.show()

print('clean border dimension:', np.shape(clean_border))
print('clean border data type:', clean_border.dtype)
```

extract fluorescent cell with disk type filter

```python
radius=20
se1 = skimage.morphology.disk(radius)
print("dim sel:", np.shape(se1))
plt.imshow(se1,'gray')
plt.show()

imflat1 = skimage.morphology.white_tophat(imflatgray, selem= se1)
print("scale imflat1:", imflat1.max(), imflat1.min())

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(16,16), sharex=True, sharey=True)
ax1.imshow(imflatgray, 'gray')
ax1.set_title('gray scale', fontsize=10)
ax2.imshow(imflat1, 'gray')
ax2.set_title('filtered with disk radius 20', fontsize=10)
plt.show()
```

```python
imflat1 = skimage.morphology.white_tophat(imflatgray, selem= se1)
print('scale for imflat passing a fliter with radius=%d :' % (radius))
print('imflat scale:',imflat1.max(), imflat1.min())

#th = 130
th = 0.02
imflat1_mask = imflat1 > th

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Binary image')
ax.axis('off')
ax.imshow(imflat1_mask, cmap='gray')
plt.show()

imflat1 = imflat1_mask
```

```python
se2 = skimage.morphology.rectangle(6,1)
imflat2 = skimage.morphology.opening(imflat1, selem=se2)
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('noise removal')
ax.axis('off')
tmp = ax.imshow(imflat2, cmap='gray')
```

```python
se2 = skimage.morphology.disk(radius =5.5)
imflat3 = skimage.morphology.closing(imflat2, selem=se2)
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('noise removal')
ax.axis('off')
tmp = ax.imshow(imflat3, cmap='gray')
```

```python
se2 = skimage.morphology.disk(radius = 3)
imflat4 = skimage.morphology.erosion(imflat3, selem=se2)
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('noise removal')
ax.axis('off')
tmp = ax.imshow(imflat4, cmap='gray')
```

```python
d = scipy.ndimage.distance_transform_edt(imflat2)
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Distance image')
ax.axis('off')
tmp = ax.imshow(d, cmap='gray', clim=(0, 10))
```

```python
# Convert from peaks to troughs for watershed

d = -d
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Distance image inverted')
ax.axis('off')
tmp = ax.imshow(d, cmap='gray')
```

```python
h = 2
markers = skimage.morphology.h_minima(d, h)
se = skimage.morphology.disk(radius=2)

markersBig = skimage.morphology.dilation(markers, se)


fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Markers')
ax.axis('off')
tmp = ax.imshow(markersBig, cmap='gray')
```

```python
# Run a watershed on the distance matrix

se3 = skimage.morphology.disk(radius=2)

labeledMarkers, numFeatures = scipy.ndimage.measurements.label(markersBig)
print('Number of nuclei: %d' % numFeatures)

w = skimage.morphology.watershed(d, labeledMarkers, watershed_line=True)
# , mask=bw

# This just "widens" the erosion lines to more clearly separate objects.
    
w = skimage.morphology.erosion(w, selem=se3)

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Watershed image')
ax.axis('off')
tmp = ax.imshow(w, cmap='gray', clim=(0,1))
```

```python
bwcut = np.logical_and(imflat2, w)  # if both pixel values are 1, give back a 1, else assign 0
bwcut = skimage.segmentation.clear_border(bwcut)

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Binary image with cuts from watershed')
ax.axis('off')
tmp = ax.imshow(bwcut, cmap='gray')
```

```python
bwcut = np.logical_and(imflat4, w)  # if both pixel values are 1, give back a 1, else assign 0
    
bwcut = skimage.segmentation.clear_border(bwcut)
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Binary image with cuts from watershed')
ax.axis('off')
tmp = ax.imshow(bwcut, cmap='gray')
```

```python
bwLabels, numFeatures = skimage.morphology.label(bwcut, return_num=True)
print('Number of nuclei: %d' % numFeatures)
print('dim bwLabels:', np.shape(bwLabels))

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title('Labeled image')
ax.axis('off')
tmp = ax.imshow(bwLabels, cmap='gray', clim=(0, numFeatures))
```

```python
# get the position of each nucleus and centroid of the whole embryo
props = skimage.measure.regionprops(bwLabels, intensity_image=imflatgray)
xpos = []
ypos = []
bcd_conc = []
for prop in props:
    xpos.append(prop.centroid[0])
    ypos.append(prop.centroid[1])
    bcd_conc.append(prop.mean_intensity)

xmean = np.mean(xpos)
ymean = np.mean(ypos)

xpos_norm = xpos - xmean
ypos_norm = ypos - ymean


theta = np.arctan2(xpos_norm,ypos_norm)

plt.plot(theta,bcd_conc,"b.")
plt.xlabel("Nucleus position")
plt.ylabel("Erk intensity(AU)" )
plt.show()


```
