```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

```python
def getintensity(positions,genename):
    total_x=[]
    total_y=[]
    for p in positions:
        df=pd.read_csv('%s-%s.csv'%(genename,p), sep=',',header=0,encoding="utf8")
        x = df.values[:,0]
        y = df.values[:,1]
        x_norm = (x/x[-1])*100 # normalize to EL %
        y_filtered=signal.savgol_filter(y, 101, 3) # filter out signal noise by Savitzky–Golay filter
        y_norm = (y_filtered-min(y_filtered))/(max(y_filtered)-min(y_filtered)) #normalize by min max values
        total_x.extend(x_norm)
        total_y.extend(y_norm)
        plt.plot(x_norm,y_norm)
        plt.show
    
    plt.xlabel("Egg lengh %")
    plt.ylabel("Intensity")
    plt.title("%s Intensity of uBcd posteriorly illuminated" %genename)
```

```python
def getrawintensity(positions,genename):
    total_x=[]
    total_y=[]
    for p in positions:
        df=pd.read_csv('%s-%s.csv'%(genename,p), sep=',',header=0,encoding="utf8")
        x = df.values[:,0]
        y = df.values[:,1]
        x_norm = (x/x[-1])*100 # normalize to EL %
        y_filtered=signal.savgol_filter(y, 101, 3) # filter out signal noise by Savitzky–Golay filter
    
        total_x.extend(x_norm)
        total_y.extend(y_filtered)
        plt.plot(x_norm,y_filtered)
        plt.show
    
    plt.xlabel("Egg lengh %")
    plt.ylabel("Intensity")
    plt.title("%s Intensity of uBcd posteriorly illuminated" %genename)
```

```python
positions = ["7","8","12","15","16","20","21","22","23","24","27","37","38","44","46","51"]
genename = "Btd"
getrawintensity(positions,genename)
```
